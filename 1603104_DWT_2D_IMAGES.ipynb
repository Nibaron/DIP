**A discrete wavelet transform (DWT) is a transform that decomposes a given signal into a number of sets, where each set is a time series of coefficients describing the time evolution of the signal in the corresponding frequency band.
it transforms can be used for image processing. As the resolution of the image
increases, it requires a lot of disk space. DWT is used to reduce the size of an imagewithout compromising on quality and hence resolution increases.**

```python
from PIL import Image
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import ndimage
import cv2
import pywt
import pywt.data

I = cv2.imread('/content/drive/MyDrive/DIP/person.png', 0)
```

**The Real Image has been shown here.**

```python
plt.figure(figsize=(30,30))
plt.show()

plt.imshow(I,interpolation="nearest",cmap=plt.cm.gray)
plt.title('original image',fontsize=10)
```

**Here, we have created approximation band ,horizontal band ,vertical band and diagonal band of the original image.**

```python
coeffs2 = pywt.dwt2(I, 'bior1.3')
LL, (LH, HL, HH) = coeffs2
fig = plt.figure(figsize=(12, 3))
titles = ['Approximation', ' Horizontal detail',
      'Vertical detail', 'Diagonal detail']
for i, a in enumerate([LL, LH, HL, HH]):
  ax = fig.add_subplot(1, 4, i + 1)
  ax.imshow(a, interpolation="nearest", cmap=plt.cm.gray)
  ax.set_title(titles[i], fontsize=10)
  ax.set_xticks([])
  ax.set_yticks([])

fig.tight_layout()
plt.show()
```

**Using the four bands from above, we have created the Reconstraction images here.**

```python
rec=pywt.idwt2(coeffs2,'db3')
rec=np.uint8(rec)

plt.figure(figsize=(30,30))
plt.show()

plt.imshow(rec,interpolation="nearest",cmap=plt.cm.gray)
plt.title('reconstruction image',fontsize=10)
```
